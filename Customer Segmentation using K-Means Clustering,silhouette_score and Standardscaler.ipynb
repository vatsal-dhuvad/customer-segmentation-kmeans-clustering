##Customer Segmentation using K-Means Clustering,silhouette_score and Standardscaler##

```python
#Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
customer_data = pd.read_csv('Mall_Customers.csv')#Load dataset
customer_data.head() #first five rows
```

```python
X = customer_data.iloc[:,[3,4]].values #Choosing the Annual Income Column & Spending Score column
print(X)
```

```python
#Choosing the number of clusters
# finding WCSS(Within Clusters Sum of Squares) value for different number of clusters
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)
```

```python
# plot an elbow graph

sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
print("\n=== Training Advanced Multi-Dimensional Model ===")

# Extract target data columns
X_advanced = customer_data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].copy()

# Convert categories to numbers
X_advanced['Gender'] = customer_data['Gender'].map({'Female': 0, 'Male': 1})

# Normalize different feature units
scaler = StandardScaler()
X_advanced_scaled = scaler.fit_transform(X_advanced)

# Initialize clustering algorithm settings
kmeans_adv = KMeans(n_clusters=5, init='k-means++', random_state=42)
# Assign groups to customers
Y_advanced = kmeans_adv.fit_predict(X_advanced_scaled)

# Save group labels back
customer_data['Advanced_Cluster'] = Y_advanced

# Measure cluster separation quality
silhouette_avg_4d = silhouette_score(X_advanced_scaled, Y_advanced)
print(f"[4D Model] Average Silhouette Score: {silhouette_avg_4d:.4f}")
```

```python
# Add the cluster labels back to the original DataFrame
customer_data['Cluster_Group'] = Y

# Group by the cluster to find the mean Age, Income, and Spending Score
cluster_profiles = customer_data.groupby('Cluster_Group')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
cluster_profiles['Cluster_Size'] = customer_data['Cluster_Group'].value_counts()

print("\n--- Cluster Profiles ---")
display(cluster_profiles)
```
